```python
import spacy
from spacy.matcher import PhraseMatcher
from spellchecker import SpellChecker  # Importing spell checker for handling spelling errors
import pandas as pd

# Load spaCy's pre-trained English model
nlp = spacy.load('en_core_web_md')

# Initialize the PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)

# List of key phrases we want to match
patterns = ["admission process", "application fee", "pre-enrollment assessment",
            "how to apply", "deadline for application"]

# Convert these patterns to spaCy Doc objects and add them to the matcher
phrase_patterns = [nlp(text) for text in patterns]
matcher.add("FAQ_PATTERNS", phrase_patterns)

# Original nested FAQ data
faq_data = {
    "Admissions": [
        {"question": "What is the process for admission into Saras AI Institute?",
         "answer": "The admission process involves submitting the online application..."},
        {"question": "Is there an application fee for applying to Saras AI Institute?",
         "answer": "There is no application fee for applying to Saras AI Institute."},
        {"question": "What is pre-enrollment assessment test? How do I prepare for it?",
         "answer": "The pre-enrollment test evaluates your personal traits and communication skills..."},
        {"question": "When is the deadline for submitting the application?",
         "answer": "The deadline for submitting applications is 5th August 2024."}
    ],
    "Curriculum and Faculty": [
        {"question": "What is the curriculum like at Saras AI Institute?",
         "answer": "The curriculum is designed to provide essential AI skills."},
        {"question": "Do you also conduct LIVE sessions?",
         "answer": "Yes, live sessions are regularly conducted to supplement recorded content."}
    ]
    # Add more categories if needed
}

# Flatten the FAQ data into a pandas DataFrame
faq_list = []
for category, faqs in faq_data.items():
    for faq in faqs:
        faq_list.append({'Category': category, 'Question': faq['question'], 'Answer': faq['answer']})

faq_df = pd.DataFrame(faq_list)

# Initialize spell checker
spell = SpellChecker()

# Function to check and correct spelling mistakes in a query
def correct_spelling(user_query):
    words = user_query.split()
    corrected_words = []

    for word in words:
        corrected_word = spell.correction(word)  # Get the most likely correction
        corrected_words.append(corrected_word)

    corrected_query = " ".join(corrected_words)
    return corrected_query

# Function to find exact phrase matches in the user query
def find_phrase_match(user_query_doc, faq_questions):
    matches = matcher(user_query_doc)
    matched_phrases = [user_query_doc[start:end].text.lower() for match_id, start, end in matches]

    # Check if any of the FAQ questions contain the matched phrases
    for question in faq_questions:
        question_doc = nlp(question)
        for phrase in matched_phrases:
            if phrase in question_doc.text.lower():
                return question
    return None

# Function to find the best match using similarity scoring
def find_best_match(user_query_doc, faq_df):
    faq_df['Question_doc'] = faq_df['Question'].apply(nlp)
    faq_df['Similarity'] = faq_df['Question_doc'].apply(lambda x: x.similarity(user_query_doc))
    best_match_index = faq_df['Similarity'].idxmax()
    best_match_question = faq_df.loc[best_match_index, 'Question']
    best_match_answer = faq_df.loc[best_match_index, 'Answer']
    return best_match_question, best_match_answer, faq_df['Similarity']

# Main function to handle both phrase matching and fallback to similarity scoring
def get_best_faq_response(user_query, faq_df):
    # Correct spelling errors in the user query
    corrected_query = correct_spelling(user_query)
    print(f"Corrected Query: {corrected_query}")

    user_query_doc = nlp(corrected_query)

    # Try to find phrase matches
    matching_faq = find_phrase_match(user_query_doc, faq_df['Question'])

    if matching_faq:
        # Return the exact match answer
        matching_answer = faq_df.loc[faq_df['Question'] == matching_faq, 'Answer'].values[0]
        return matching_faq, matching_answer, faq_df
    else:
        # Fallback to best similarity match
        best_match_question, best_match_answer, similarities = find_best_match(user_query_doc, faq_df)
        best_similarity = similarities.max()

        if 0.60 <= best_similarity <= 0.75:
            # Print the result and add the question-answer pair to the DataFrame
            print(f"Best match found with {best_similarity * 100:.2f}% similarity.")
            faq_df = faq_df.append({'Category': 'Unknown', 'Question': user_query, 'Answer': best_match_answer}, ignore_index=True)
            return best_match_question, best_match_answer, faq_df
        else:
            return best_match_question, best_match_answer, faq_df

# Example user query with a spelling mistake
user_query = "What is the admission process"

# Get the best FAQ response
best_question, best_answer, updated_faq_df = get_best_faq_response(user_query, faq_df)

print(f"Best Match Question: {best_question}")
print(f"Answer: {best_answer}")

# Display the updated DataFrame
print("\nUpdated FAQ DataFrame:")
print(updated_faq_df)
```

```python
import spacy
from spacy.matcher import PhraseMatcher
from spellchecker import SpellChecker  # Importing spell checker for handling spelling errors
import pandas as pd

# Load spaCy's pre-trained English model
nlp = spacy.load('en_core_web_md')

# Initialize the PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)

# List of key phrases we want to match
patterns = ["admission process", "application fee", "pre-enrollment assessment",
            "how to apply", "deadline for application"]

# Convert these patterns to spaCy Doc objects and add them to the matcher
phrase_patterns = [nlp(text) for text in patterns]
matcher.add("FAQ_PATTERNS", phrase_patterns)

# Original nested FAQ data
faq_data = {
    "Admissions": [
        {"question": "What is the process for admission into Saras AI Institute?",
         "answer": "The admission process involves submitting the online application..."},
        {"question": "Is there an application fee for applying to Saras AI Institute?",
         "answer": "There is no application fee for applying to Saras AI Institute."},
        {"question": "What is pre-enrollment assessment test? How do I prepare for it?",
         "answer": "The pre-enrollment test evaluates your personal traits and communication skills..."},
        {"question": "When is the deadline for submitting the application?",
         "answer": "The deadline for submitting applications is 5th August 2024."}
    ],
    "Curriculum and Faculty": [
        {"question": "What is the curriculum like at Saras AI Institute?",
         "answer": "The curriculum is designed to provide essential AI skills."},
        {"question": "Do you also conduct LIVE sessions?",
         "answer": "Yes, live sessions are regularly conducted to supplement recorded content."}
    ]
    # Add more categories if needed
}

# Flatten the FAQ data into a pandas DataFrame
faq_list = []
for category, faqs in faq_data.items():
    for faq in faqs:
        faq_list.append({'Category': category, 'Question': faq['question'], 'Answer': faq['answer']})

faq_df = pd.DataFrame(faq_list)

# Initialize spell checker
spell = SpellChecker()

# Function to check and correct spelling mistakes in a query
def correct_spelling(user_query):
    words = user_query.split()
    corrected_words = []

    for word in words:
        corrected_word = spell.correction(word)  # Get the most likely correction
        corrected_words.append(corrected_word)

    corrected_query = " ".join(corrected_words)
    return corrected_query

# Function to find exact phrase matches in the user query
def find_phrase_match(user_query_doc, faq_questions):
    matches = matcher(user_query_doc)
    matched_phrases = [user_query_doc[start:end].text.lower() for match_id, start, end in matches]

    # Check if any of the FAQ questions contain the matched phrases
    for question in faq_questions:
        question_doc = nlp(question)
        for phrase in matched_phrases:
            if phrase in question_doc.text.lower():
                return question
    return None

# Function to find the best match using similarity scoring
def find_best_match(user_query_doc, faq_df):
    faq_df['Question_doc'] = faq_df['Question'].apply(nlp)
    faq_df['Similarity'] = faq_df['Question_doc'].apply(lambda x: x.similarity(user_query_doc))
    best_match_index = faq_df['Similarity'].idxmax()
    best_match_question = faq_df.loc[best_match_index, 'Question']
    best_match_answer = faq_df.loc[best_match_index, 'Answer']
    return best_match_question, best_match_answer, faq_df['Similarity']

# Main function to handle both phrase matching and fallback to similarity scoring
def get_best_faq_response(user_query, faq_df):
    # Correct spelling errors in the user query
    corrected_query = correct_spelling(user_query)
    print(f"Corrected Query: {corrected_query}")

    user_query_doc = nlp(corrected_query)

    # Try to find phrase matches
    matching_faq = find_phrase_match(user_query_doc, faq_df['Question'])

    if matching_faq:
        # Return the exact match answer
        matching_answer = faq_df.loc[faq_df['Question'] == matching_faq, 'Answer'].values[0]
        return matching_faq, matching_answer, faq_df
    else:
        # Fallback to best similarity match
        best_match_question, best_match_answer, similarities = find_best_match(user_query_doc, faq_df)
        best_similarity = similarities.max()

        if 0.60 <= best_similarity <= 1:
            # Print the result and add the question-answer pair to the DataFrame
            print(f"Best match found with {best_similarity * 100:.2f}% similarity.")
            return best_match_question, best_match_answer, faq_df
        else:
            return best_match_question, best_match_answer, faq_df

# Example user query with a spelling mistake
user_query = "When do I need to submit my application by?"
# Get the best FAQ response
best_question, best_answer, updated_faq_df = get_best_faq_response(user_query, faq_df)

print(f"Best Match Question: {best_question}")
print(f"Answer: {best_answer}")

# Display the updated DataFrame
print("\nUpdated FAQ DataFrame:")
print(updated_faq_df)
```

```python
import spacy

# Load a spaCy model with word vectors
nlp = spacy.load("en_core_web_md")

# Define two sentences
sentence1 = "When do I need to submit my application by?"
sentence2 = "What is pre-enrollment assessment test? How do I prepare for it?"

# Process the sentences using spaCy
doc1 = nlp(sentence1)
doc2 = nlp(sentence2)

# Calculate cosine similarity
similarity_score = doc1.similarity(doc2)

print(f"Cosine Similarity between the sentences: {similarity_score:.2f}")
```
